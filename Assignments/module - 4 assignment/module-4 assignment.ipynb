You are provided with the following lines of codes about the employees of an organization. Assuming each of these employees is required to make a one-off donation equivalent to their annual base salary divided by the age (in years), multiplied by the years in the current position. Can you calculate how much donation each person will make and add this as a new column to the dataframe?

```python
# helper code (do not change or delete)

# importing libraries
import pandas as pd
import numpy as np

np.random.seed(1)

# parameters
names = ["Adekunle", 
         "Saheed", 
         "Fahad", 
         "Fatimah", 
         "Sheriff", 
         "Abdul", 
         "Daniel", 
         "Simon", 
         "Abiodun", 
         "Ammar"]

ages = np.random.choice(np.arange(20, 51, 1), 
                        size = len(names)
                       )

job_title = ["lecturer",
             "developer", 
             "researcher", 
             "engineer", 
             "manager", 
             "lawyer", 
             "builder", 
             "data scientist", 
             "professor",
             "geoscientist"]

# defining the dataframe
df = pd.DataFrame({"name": names, 
                   "age": ages, 
                   "position": np.random.choice(job_title, len(names), replace = False),
                   "years in position": np.random.randint(1, 10, size = len(names)),
                   "annual_base_salary": np.abs(np.random.normal(size = (len(names)))*1000000)
                  })

# manipulating the dataframe
df.annual_base_salary = df.annual_base_salary.round(-3).astype(int)
df["bonus"] = df.apply(lambda x: x["years in position"]*x["annual_base_salary"]*0.1, axis = 1)
df["bonus"] = df["bonus"].round(-3).astype(int)

# taking a look at the dataframe
df
```
